# Análise de Vendas de Lojas de Tecnologia

Projeto de análise exploratória de dados utilizando Python, Pandas e Matplotlib
para identificar desempenho de lojas, produtos mais vendidos e métricas
estatísticas de faturamento.


## Contexto de Negócio

Uma empresa com três lojas deseja entender:

- Qual loja gera maior faturamento
- Quais produtos possuem maior volume de vendas
- Como estão as métricas estatísticas do faturamento

A análise foi realizada utilizando Python para apoiar decisões estratégicas.


```python
import pandas as pd
import matplotlib.pyplot as plt

dados = {
    "loja": ["A", "A", "B", "B", "C", "C", "A", "B", "C", "A"],
    "produto": ["Teclado", "Mouse", "Teclado", "Monitor", "Mouse", "Monitor", "Monitor", "Mouse", "Teclado", "Mouse"],
    "quantidade": [10, 20, 15, 5, 25, 7, 3, 18, 12, 30],
    "preco_unitario": [100, 50, 100, 800, 50, 800, 800, 50, 100, 50]
}

df = pd.DataFrame(dados)
df["faturamento"] = df["quantidade"] * df["preco_unitario"]

faturamento_total = df["faturamento"].sum()
faturamento_media = df["faturamento"].mean()
desvio = round(df["faturamento"].std(), 2)
faturamento_loja = df.groupby('loja')['faturamento'].sum()
faturamento_produto = df.groupby('produto')['faturamento'].sum()

print(f'Foi identificado um desvio de {desvio} na média de faturamento')

# Criar e salvar gráfico de faturamento por loja
plt.figure()
faturamento_loja.plot(kind='bar', title='Faturamento por Loja')
plt.ylabel('Faturamento')
plt.savefig('faturamento_loja.png')  # salva a imagem

# Criar e salvar gráfico de faturamento por produto
plt.figure()
faturamento_produto.plot(kind='bar', title='Faturamento por Produto')
plt.ylabel('Faturamento')
plt.savefig('faturamento_produto.png')  # salva a imagem

plt.show()

```

O produto com maior faturamento foi o Monitor, conforme demonstrado no gráfico de comparação entre os produtos.

Constatou-se que a loja C apresentou a maior quantidade de produtos vendidos em comparação às lojas A e B.
